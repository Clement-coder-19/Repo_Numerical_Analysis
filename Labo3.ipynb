```python
import math


def f(x):
  return x**2-value

// derivative of f(x)
def g(x):
  return 2*x

def newton(x0, e, N):
  step=1
  condition=True
  while condition:
    if g(x0)==0.0:
      print('Divide by zero error!')
      break
  x1 = x0 - f(x0)/g(x0)
  print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
  x0 = x1
  step += 1
  if step > N:
    flag = 0
    break
  condition = abs(f(x1)) > e
  if flag==1:
    print('\nRequired root is: %0.8f' % x1)
  else:
    print('\nNot Convergent.')

  // comparison with the true value
  true_val=math.sqrt(value)
  if abs(x1- true_val)<e:
    print("The result is in the tolerance")
  elif x1> true_val:
    print("Overestimation")
  else:
    print("Underestimation")

# Input section
value= float(input("Enter the number to compute its square root:"))
x0 = float(input('Enter Guess: '))
e = float(input('Tolerable Error: '))
N = int(input('Maximum Step: '))
newton(x0,e,N)
```
