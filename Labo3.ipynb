```python
# With Babylonian method
def babylonian_sqrt(S, x, e, N):
    for step in range(N):
        if x == 0:
            print("Division by zero error!")
            return
        x_next=0.5*(x+S/x)
        print("Iteration %d, x = %0.6f"%(step+1,x_next))
        if abs(x_next-x)<e:
            x=x_next
            break
        x=x_next

    print("\nApproximate root: x:%0.6f" %(x))
    #comparison with the true value
    if x*x>S:
      print("Overestimation")
    elif x*x< S:
      print("Underestimation")
    else:
      print("Equal")



#Input
S= float(input("Enter the number to compute its square root:"))
x0 = float(input('Enter Guess: '))
e = float(input('Tolerable Error: '))
N = int(input('Maximum Step: '))

babylonian_sqrt(S, x0, e, N)
```
